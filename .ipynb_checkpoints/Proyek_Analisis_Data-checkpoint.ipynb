## Data Wrangling

### Gathering Data

```python
dfs = dict()

dfs["aotizhongxin"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Aotizhongxin_20130301-20170228.csv")
dfs["changping"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Changping_20130301-20170228.csv")
dfs["dingling"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Dingling_20130301-20170228.csv")
dfs["dongsi"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Dongsi_20130301-20170228.csv")
dfs["guanyuan"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Guanyuan_20130301-20170228.csv")
dfs["gucheng"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Gucheng_20130301-20170228.csv")
dfs["huairou"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Huairou_20130301-20170228.csv")
dfs["nongzhanguan"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Nongzhanguan_20130301-20170228.csv")
dfs["shunyi"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Shunyi_20130301-20170228.csv")
dfs["tiantan"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Tiantan_20130301-20170228.csv")
dfs["wanliu"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Wanliu_20130301-20170228.csv")
dfs["wanshouxigong"] = pd.read_csv("./Air-quality-dataset/PRSA_Data_20130301-20170228/PRSA_Data_Wanshouxigong_20130301-20170228.csv")
```

### Assessing Data

#### Data Info

```python
for key in dfs:
    print(f"=== INFO: {str.upper(key)} ===")
    print(dfs[key].info())
    print()
```

#### Panjang Data

```python
for key in dfs:
    print(f"{key}  \t:{len(dfs[key])} baris")
```

#### Data Null

```python
pd.DataFrame([df.isna().sum() for df in dfs.values()])
```

```python
plt.figure(figsize=(10, 30)) 

for i, key in enumerate(dfs):
    plt.subplot(6, 2, i + 1)
    sns.heatmap(dfs[key].isna(), cbar=False)
```

#### Duplikasi

```python
for key in dfs:
    print(f"{key}  \t:{(dfs[key].duplicated().sum())} duplikasi")
```
