### Cleaning Data

#### Filling Missing Values

Karena data kualitas udara ini merupakan data sensor dan data yang presentasikan merupakan data per waktu, maka dapat disebut sebagai data time series, sehingga menanganan missing value dapat dilakukan dengan interpolasi untuk kolom bertipe numerik. Untuk data bertipe kategorikal dapat dilakukan dengan pengisian value dengan modus pada data kolom tersebut, yaitu kolom wind direction (wd)

```python
for key in dfs:
    dfs[key] = dfs[key].interpolate(method="linear", limit_direction="both")
    dfs[key]["wd"] = dfs[key]["wd"].fillna(dfs[key]["wd"].mode()[0])
```

Setelah dilakukan pengisian missing value, maka jumlah data yang kosong tidak ada lagi

```python
for df in dfs.values():
    print(f"Jumlah missing value: {df.isna().sum().sum()}")
```

#### Ubah Beberapa Kolom Tanggal Menjadi Satu Kolom

```python
for key in dfs:
    dfs[key]["date"] = pd.to_datetime(dfs[key][['year', 'month', 'day']])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Berapa rata-rata kadar PM2.5 dan PM10 di jam yang sama setiap harinya?


```python
day_mean_df = df.groupby(by="hour").agg({
    "PM2.5": "mean",
    "PM10": "mean",
})

day_mean_df.plot()
```

Grafik di atas menunjukkan kadar rata-rata per jam nya dalam beberapa tahun.
- Pada malam hari kadar PM2.5 maupun PM10 memiliki rata-rata yang tinggi dibandingkan pada siang hari.
- Kadar PM2.5 pada malam hari memiliki angka antara 100 hingga 120 lebih.
- Kadar PM2.5 pada siang hari memiliki angka antara 90 hingga 100 lebih.
- Kadar PM10 pada malam hari memiliki angka antara 80 hingga 90.
- Kadar PM10 pada siang hari memiliki angka di bawah 80.

### Pertanyaan 2: Station apa yang memiliki PM2.5 paling buruk pada bulan terakhir?

```python
last_year_df = df.loc[(df["year"] == df["year"].max())]
last_month_df = last_year_df.loc[(last_year_df["month"] == last_year_df["month"].max())]

last_month_df.groupby(by="station").agg({
    "PM2.5": "mean",
}).sort_values(by="PM2.5").plot(kind="bar")
```

- Didapatkan pada bulan terakhir pada tahun 2017, Wanshouxigong merupakan station dengan kualitas PM2.5 tertinggi sehingga dapat dikategorikan sebagai PM2.5 terburuk.
- Peringkat kedua adalah Nongzhanguan
- Peringkat ketiga adalah Dongsi
- Peringkat terakhir adalah Dingling
